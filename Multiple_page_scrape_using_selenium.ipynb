```python
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
url='https://www.daraz.pk/laptops/?page=2'
```

```python
driver=webdriver.Chrome()
driver.get(url)
```

```python
soup=BeautifulSoup(driver.page_source,"html.parser")
```

```python
print(soup.prettify)
```

```python
name=soup.find("div",{"class":"RfADt"}).text.strip()
```

```python
print(name)
```

```python
title=[]
for i in range(1,6):
    driver.get("https://www.daraz.pk/laptops/?page={}".format(i))
    for name in soup.find_all("div",{"class":"RfADt"}):
        tit = name.get_text(strip=True)
        title.append(tit)
```

```python
for i in title:
    print(i)
```

```python
len(title)
```

```python
Price=[]
for i in range(1,6):
    driver.get("https://www.daraz.pk/laptops/?page={}".format(i))
    for price in soup.find_all("span",{"class":"ooOxS"}):
        Pri = price.get_text(strip=True)
        Price.append(pri)
```

```python
Price
```

```python
len(Price)
```

```python
data={"Laptop_names":title,
     "Price":Price}
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(data=data)
```
